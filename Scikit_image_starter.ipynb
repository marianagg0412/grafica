#Imports

```python
from skimage import data, color
import matplotlib.pyplot as plt
import numpy as np
import skimage.util as util
from skimage.filters import gaussian, median
```

#Using data from scikit image

```python
image = data.astronaut()
plt.imshow(image)
```

```python
image.shape #x, y, RGB
```

#RGB to gray

```python
plt.imshow(image[:, :, 2], cmap='gray') #el aporte de cada color, the more aporte, the clearer
```

```python
gray_astronaut = color.rgb2gray(image) #Literally RGB to colorgray
plt.imshow(gray_astronaut, cmap='gray')
```

**Gray to RGB **
Information is lost during RGB to gray conversion

#Image transformations

```python
def compare_images(image1, image2):
  fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
  ax1.imshow(image1, cmap='gray')
  ax2.imshow(image2, cmap='gray')
  plt.show()
```

```python
#Numpy use
#Vertical flip
verticap_flip_astronaut = np.flipud(gray_astronaut)
plt.imshow(verticap_flip_astronaut, cmap='gray')
```

```python
#Vertical flip
horizontal_flip_astronaut = np.fliplr(gray_astronaut)
plt.imshow(horizontal_flip_astronaut, cmap='gray')
```

##Ruido y filtro: gaussiano

```python
image_with_noise_gaussian = util.random_noise(gray_astronaut, mode='gaussian')
plt.imshow(image_with_noise_gaussian, cmap='gray')
```

```python
smoothed_image_gaussian = gaussian(image_with_noise_gaussian, sigma=1.4)
plt.imshow(smoothed_image_gaussian, cmap='gray')
```

##Ruido: sal y pimienta, Filtro: mediana

```python
image_with_noise_sp = util.random_noise(gray_astronaut, mode='s&p')
plt.imshow(image_with_noise_sp, cmap='gray')
```

```python
smoothed_image_median = median(image_with_noise_sp, footprint=np.ones((3, 3))) #Neighborhood size = mask
plt.imshow(smoothed_image_median, cmap='gray')
```

###Comparacin de imagenes: ruido con imagen suavizada

```python
compare_images(image_with_noise_gaussian, smoothed_image_gaussian)
```

###Comparacin de imagenes: ruido S&P filtro mediana

```python
compare_images(image_with_noise_sp, smoothed_image_median)
```
